```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model = AutoModelForCausalLM.from_pretrained(
        "jinaai/starcoder-1b-textbook", device_map='auto'
    )

tokenizer = AutoTokenizer.from_pretrained("jinaai/starcoder-1b-textbook")
```

```python
prompt = '''
Given n number of integers as input return the sum of the integers as output

def add_integers
'''

inputs = tokenizer(prompt.rstrip(), return_tensors="pt").to("cuda")

generation_output = model.generate(
    **inputs,
    max_new_tokens=256,
    eos_token_id=tokenizer.eos_token_id,
    return_dict_in_generate=True,
)

s = generation_output.sequences[0]
output = tokenizer.decode(s, skip_special_tokens=True)

print("Model Output:")
print(output)
```

Training model with my custom dataset generated to use the model in custom application

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, TextDataset, DataCollatorForLanguageModeling, Trainer, TrainingArguments
```

```python
# Load the model and tokenizer
model = AutoModelForCausalLM.from_pretrained("jinaai/starcoder-1b-textbook")
tokenizer = AutoTokenizer.from_pretrained("jinaai/starcoder-1b-textbook")
```

```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page_num in range(len(reader.pages)):
            text += reader.pages[page_num].extract_text()
    return text


def save_text_to_txt(text, output_path):
    with open(output_path, 'w', encoding='utf-8') as file:
        file.write(text)

text = extract_text_from_pdf('/content/mental_health')
txt_output_path = "output_filename.txt"
save_text_to_txt(text, txt_output_path)


```

```python
# Tokenize the dataset
train_dataset = TextDataset(
    tokenizer=tokenizer,
    file_path='output_filename.txt',  # replace with your file path
    block_size=128
)
eval_dataset = TextDataset(
    tokenizer=tokenizer,
    file_path='output_filename.txt',   # replace with your file path
    block_size=128
)

# Define data collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False
)
```

```python


# Define training arguments
training_args = TrainingArguments(
    output_dir="./results",
    overwrite_output_dir=True,
    num_train_epochs=1,  # You can adjust the number of epochs
    per_device_train_batch_size=4,
    save_steps=10_000,
    save_total_limit=2,
)

# Create Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)

# Train the model
trainer.train()

# Save the model and tokenizer
model.save_pretrained("./results")
tokenizer.save_pretrained("./results")
```

```python
def chat_with_bot(prompt):
    input_ids = tokenizer.encode(prompt, return_tensors="pt")
    output = model.generate(input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)
    decoded_output = tokenizer.decode(output[:, input_ids.shape[-1]:][0], skip_special_tokens=True)
    return decoded_output

response = chat_with_bot("I've been feeling very anxious lately.")
print(response)
```
